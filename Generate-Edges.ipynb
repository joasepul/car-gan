```python
import cv2
import glob
import matplotlib.pyplot as plt
```

```python
path_to_dataset = './stanford-cars-dataset/cars_train/'
path_to_save_data = './edge-data/'
```

```python
filespaths = glob.glob(path_to_dataset + '00[0-9][0-9][0-9].jpg') 
```

```python
for index, file in enumerate(filespaths):
    img = cv2.imread(file, cv2.IMREAD_COLOR)
    if index % 100 == 0:
        print(index)
    edge = cv2.bitwise_not(cv2.Canny(img, 100,100, L2gradient=False))
    x,y = edge.shape
    margin = (y-x)//2
    edge = edge[:,margin:y-margin]
    img = img[:,margin:y-margin]
    edge = cv2.resize(edge, (64,64))
    img = cv2.resize(img, (64,64))
    cv2.imwrite('./edge-data/edge-'+str(index)+'.png', edge)
    cv2.imwrite('./edge-data/' +str(index)+'.png', img)
       
```
