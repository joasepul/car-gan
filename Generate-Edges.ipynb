```python
import cv2
import glob
import matplotlib.pyplot as plt
import os
```

```python
path_to_dataset = os.path.join('./','cars_train/')
path_to_save_data = os.path.join('./','edge-data/')
```

```python
filespaths = glob.glob(path_to_dataset + '0[0-9][0-9][0-9][0-9].jpg') 
filespaths
```

```python
for index, file in enumerate(filespaths):
    img = cv2.imread(file, cv2.IMREAD_COLOR)
    if index % 1000 == 0:
        print(index)
    x,y,_ = img.shape
    margin = (y-x)//2
    
    img = img[:,margin:y-margin]
    edge = cv2.bitwise_not(cv2.Canny(img, 100,100, L2gradient=False))
    edge = cv2.resize(edge, (128,128))
    img = cv2.resize(img, (128,128))
    cv2.imwrite('./edge-data/edge-'+str(index)+'.png', edge)
    cv2.imwrite('./edge-data/' +str(index)+'.png', img)
       
```
